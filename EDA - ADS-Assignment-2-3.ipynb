```python
#Loading libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn

%matplotlib inline
```

```python
#reading the dataset
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

```python
#viewing the data dimension
df.shape
```

```python
#view the columns
df.columns
```

```python
#view the data types
df.dtypes
```

```python
df.info()
```

```python
#view for any unique values
df.nunique()
```

```python
#checking for any null values
df.isnull().sum()
```

```python
#Distance form home by Job Role and Attrition
df.groupby(['JobRole','Attrition'])['DistanceFromHome'].mean().unstack()
```

```python
#Visualization of Distance form home by Job Role and Attrition
fig = plt.subplots(figsize=(8,12))
sn.violinplot(x='DistanceFromHome', y='JobRole', hue='Attrition', data=df)
```

```python
#Attrition Avaerage
Atr_count = df['Attrition'].value_counts()
Atr_count
```

```python
#Visualization
sn.countplot(df['Attrition'])
fig = plt.gcf()
fig.set_size_inches(10,10)
```

```python
#Education Avera
Edu_count = df['Education'].value_counts()
Edu_count
```

```python
#Education Visualtion
sn.countplot(df['Education'])
fig = plt.gcf()
fig.set_size_inches(10,10)
```

```python
#monthly income distribution
df['MonthlyIncome'].hist(bins=12,histtype='bar', color='indigo', rwidth=0.9)
```

```python
df.groupby(['Education','Attrition'])['MonthlyIncome'].mean().unstack()
```

```python
#Comparing Monthly Income, Education and Attrition
fig = plt.subplots(figsize=(12,9))
sn.violinplot(x='Education', y='MonthlyIncome', hue='Attrition', data=df)
```
